```python

def page_crawler(year, day, month, no_dups_set):
    corpus = []
    if day<10:
        day = "0" + str(day)
    else:
        day = str(day)
    if month<10:
        month = "0" + str(month)
    else:
        month = str(month)
    
    year = str(year)
        
    
    #base url
    url = 'https://www.lemonde.fr/archives-du-monde/'   #01-01-2021/ day-month-year
    
    #take year, month, day, to get to that day's articles
    
    #using this to change pages, not smart, but it's what i got
    page_iter_url = url
    #if we're done with the first page, start appending new info to the url
    
    page_iter_url += day+ "-" + month + "-" + year + "/"

    #find all articles for that day
    results = requests.get(page_iter_url)
    doc = BeautifulSoup(results.text, "html.parser")
    
    articles = doc.findAll("a", class_="teaser__link")
    
    for article in articles:
        #print(a.prettify())
        link = article["href"]
        art_url = link
        
        #print("---------------------------------------------------")
        #print("url: ", url)
        #print("page_iter: ", page_iter_url)
        #print("art_url: ", art_url)
        #print("---------------------------------------------------")
    
        #do our scraping fun
        results = requests.get(art_url)

        doc = BeautifulSoup(results.text, "html.parser")

        paragraphs = doc.findAll("p" ,class_="article__paragraph")

        for p in paragraphs:
            sentences = p.getText()
            if sentences in no_dups:
                print("duplicate!")
                continue
            no_dups_set.add(sentences)
            
            sentences = sentences.split(".")
            for s in sentences:
                if len(s) >4:
                    s += "."
                    #print(s.strip())
                    corpus.append(s.strip())
    return(corpus, no_dups_set)
```

```python
sentence_count = 0
day = 1
month = 1
year = 2021

corpus = []
no_dups = set()

with open("corpus_fra.txt", "w",encoding="utf-8") as f:

    counter = 0

    while sentence_count < 15000:
        print("trying to run corpus_crawler")
        corpus, no_dups= page_crawler(year, day, month, no_dups)
        print("corpus crawler done")
        sentence_count += len(corpus)
        print(sentence_count)
        for sentence in corpus:
            if sentence_count < 15000:
                #print(sentence)
                print(sentence, file=f)
        
        #now we need to go to the next page with another iteration		
        day += 1
        print(day)
        if day >25:
            day = 1
            month += 1
            if month >12:
                month = 1
                year -= 1
        
        print("sentences: ", sentence_count)
        print("date: ", str(day), "-", str(month), "-", str(year))
            
        counter += 1
        if counter >50:
            counter = 0
```
